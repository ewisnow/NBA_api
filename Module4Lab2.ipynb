```python
#Load Packages
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import json
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
```

```python
nba_teams = teams.get_teams()
```

```python
#find NBA ID for GSW
warriors = [team for team in nba_teams if team['abbreviation'] == 'GSW'][0]
warriors_id = warriors['id']
#use gamefinder to find game results using GSW ID
gamefinder_gsw = leaguegamefinder.LeagueGameFinder(team_id_nullable=warriors_id)
games_gsw = gamefinder_gsw.get_data_frames()[0]
#filter games to 19-20 season
games_gsw_1920 = games_gsw[games_gsw.SEASON_ID.str[-4:] == '2019']

knicks = [team for team in nba_teams if team['abbreviation'] == 'NYK'][0]
knicks_id = knicks['id']
gamefinder_nyk = leaguegamefinder.LeagueGameFinder(team_id_nullable=knicks_id)
games_nyk = gamefinder_nyk.get_data_frames()[0]
games_nyk_1920 = games_nyk[games_nyk.SEASON_ID.str[-4:] == '2019']

bulls = [team for team in nba_teams if team['abbreviation'] == 'CHI'][0]
bulls_id = bulls['id']
gamefinder_chi = leaguegamefinder.LeagueGameFinder(team_id_nullable=bulls_id)
games_chi = gamefinder_chi.get_data_frames()[0]
games_chi_1920 = games_chi[games_chi.SEASON_ID.str[-4:] == '2019']

#concatenate dataframes together for gsw, nyk, chi
combined_df = pd.concat([games_gsw_1920, games_nyk_1920, games_chi_1920])

#write to csv
combined_df.to_csv("nba_module4_lab2.csv", index=False)
```
